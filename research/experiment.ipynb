```python
from langchain_openai import ChatOpenAI 
```

```python
llm_ques_gen_pipeline = ChatOpenAI(
    model= "gpt-4o-mini",
    temperature= 0.3
)
```

```python
prompt_template = """ 
You are an expert at creating questions based on coding materials and documentation.
Your goal is to prepare a coder or programmer for their exam and coding tests.
You do this by asking questions about the text below:
Create Only 10 Questions that will prepare the coders or programmers for their exams and coding tests. Make sure only 10 questions are created.

--------------
{text}
--------------


Make sure not to loose any important information from the text.

Questions: 
"""
```

```python
from langchain.prompts import PromptTemplate
```

```python
PROMPT_QUESTIONS = PromptTemplate(template=prompt_template, input_variables=["text"])
```

```python
refine_template = ("""
You are an expert at creating questions based on coding materials and documentation.
Your goal is to prepare a coder or programmer for their exam and coding tests.
We have received some practice questions to a certain extent: {existing_answer}
We have the option to refine the existing questions or create new ones.
(only if necessary) with some more context below.Given the new context, refine the original questions in English. Make sure create only 10 Questions not more than that.

--------------
{text}
--------------


If the context is not helpful, please provide the original questions.

Questions:
""")
```

```python
REFINE_PROMPT_QUESTIONS = PromptTemplate(template=refine_template, input_variables=["text", "existing_answer"])
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
ques_gen_chain = load_summarize_chain(
    llm=llm_ques_gen_pipeline,
    chain_type="refine",
    verbose=True,
    question_prompt=PROMPT_QUESTIONS,
    refine_prompt=REFINE_PROMPT_QUESTIONS
)
```

```python
ques = ques_gen_chain.run(document_ans_gen)
print(ques)
```

```python
print(ques)
```

```python
vector_store = FAISS.from_documents(document_ans_gen, embeddings)
```

```python
llm_ans_gen = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.3
)
```

```python
from langchain.chains import RetrievalQA
```

```python
answer_gen_chain = RetrievalQA.from_chain_type(
    llm=llm_ans_gen,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)
```

```python
ques_list = ques.split("\n")
```

```python
ques_list = [ques for ques in ques_list if ques]
```

```python
ques_list
```

```python
for question in ques_list:
    print("Question: ", question)
    answer = answer_gen_chain.run(question)
    print("Answer: ", answer)
    print("-------------------\n\n")
    
    # Save answer to file
    with open("answers.txt", "a") as file:
        file.write("Question: " + question + "\n")
        file.write("Answer: " + answer + "\n")
        file.write("\n\n")
    
```
